# Scatterplot

```
X = np.random.randn(100,2) # 100 observations of 2 variables
```

```
# create a cross sectional plot , between 0th column and 1st column, 
# : because all horizontal axis, and 0,1 correspond to vertical axis of columns
plt.scatter(X[:,0],X[:,1])
```

```
#lets use diff colors to classify
X=np.random.randn(20,2) # all centered at zero, 2 column vector of 20 rows
X
```

```
X[:10]+=3 #select half the rows and add 3
X
```

```
Y= np.zeros(20) # 20 rows of zeros
Y
```

```
Y[:10]=1 #first ten go blue
Y #set half of these to 1, to change coloring scheme for these as we pass this into c variable
```

```
plt.scatter(X[:,0],X[:,1], c=Y) # c is a 1 dim list corresponding to colors 
```

# Plotting Images

```
!wget https://www.denofgeek.com/wp-content/uploads/2018/03/dragon-ball-z-super-saiyan-goku.jpg?resize=768%2C432
```

```
from PIL import Image
```

```
im = Image.open('dragon-ball-z-super-saiyan-goku.jpg?resize=768,432')
```

```
type(im)
```

```
arr = np.array(im) #images are represented as 2d arrays
```

```
arr
```

```
#all these are numbers between 0 and 255 and are all ints
# 2 to 8 bits, or 256 bits
arr.shape #first two are spatial dims height and width
#3rd dim is color stored using 3 values (rgb)
```

```
plt.imshow(arr);
```

```
# convert to grayscale
#take mean across color channels
gray= arr.mean(axis=2)
gray
# you are flattenning rgb into one average of r and g and b
#https://stackoverflow.com/questions/64545130/rgb-to-grayscale-average-method-python
```

```
gray.shape #color dimension collapsed
```

```
plt.imshow(gray); #why green and yello 
# all numbers between 0 and 255, decided by matplot lib
# might come across diff in diff languages
# this is indeed a heatmap
```

```
plt.imshow(gray,cmap='gray'); #enforce a color cmap
```

```
plt.imshow(gray,cmap='pink'); #enforce a diff color cmap
# from https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
```

```
plt.imshow(arr,cmap='autumn'); #on original image cmap has no effect
```

```
plt.imshow(gray,cmap='autumn'); # enforces autumn on gray image
```

```
#x1 or x2 are 1, y is 1
#else y is zero
N=1000
X1=np.random.rand(N) *2 -1
X2=np.random.rand(N) *2 -1
```

```
plt.hist(X1);
```

```
plt.hist(X2);
```

```
X1.mean()
```

```
X2.mean()
```

```
Y=np.zeros(N)
for i in range(N):
  if X1[i] * X2[i] <0:
    Y[i]=1
```

```
plt.scatter(X1,X2, c=Y);
```
